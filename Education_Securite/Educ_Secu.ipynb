```python
# regarder diff de prix pour meme superificie entre maison et appartement

import pandas as pd
import requests
import geopandas as gpd
from shapely.geometry import Point
pd.set_option("display.max_columns", None)

url1 = r'https://raw.githubusercontent.com/klopstock-dviz/immo_vis/master/data/ech_annonces_ventes_68.csv'
url2 = r'https://raw.githubusercontent.com/klopstock-dviz/immo_vis/master/data/ech_annonces_locations_68.csv'

df_ventes = pd.read_csv(url1, sep=';')
df_locations = pd.read_csv(url2, sep=';')
display(df_ventes.head())


'''
Ventes:
    type_annonceur: uniquement privé, pas de pa ?
    typedebien: appart ou maison
    UU2010 : code insee de l'unité urbaine
    REG : region (remove?)
'''
print(df_ventes['type_annonceur'].value_counts())
```
